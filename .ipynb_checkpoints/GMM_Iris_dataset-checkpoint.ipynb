# GMM

```python
X = df[['petal length (cm)','petal width (cm)']].to_numpy()
y = df['Iris type'].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 3)
```

```python
def plot_2dnormal(normal, color):
    # plot center
    plt.plot(normal.mean[0], normal.mean[1], '*', ms=5, color=color)

    # get countour grid
    x, y = np.mgrid[0:7:.01, 0:3:.01]
    
    # plot countour
    pos = np.dstack((x, y)) 
    plt.contour(x, y, normal.pdf(pos), levels=7)


def e_step(X, weights, components):
    unnormalized_logprobs = [
        np.log(w) + component.logpdf(X) for w, component in zip(weights, components)
    ]
    probs_matrix = softmax(np.vstack(unnormalized_logprobs), axis=0)
    return probs_matrix


def m_step(X, probs):
    
    sum_probs = probs.sum(axis=1)
    
    weights = sum_probs / probs.shape[1]
    mus = (probs @ X) / sum_probs[:, None]
    centeredX = X - mus[:, None, :]
    probs_centeredX = probs[..., None] * centeredX
    covs = (probs_centeredX.transpose(0,2,1) @ centeredX) / sum_probs[:, None, None]
    
    components = [
        multivariate_normal(mean=mu, cov=cov) for mu, cov in zip(mus, covs)
    ]
    
    return weights, components
```

# Use GMM for prediction

```python
# Train GMM for 20 iterations
for i in range(20):    
    # expectation step
    probs = e_step(X_train, weights_estimated, components_estimated)
    # maximization step
    weights_estimated, components_estimated = m_step(X_train, probs)
```

```python
predictions = e_step(X_test, weights_estimated, components_estimated).argmax(axis=0)
```

```python
assert len(predictions) == len(y_test), "Unequal arrays"
acc = sum([p == y for p, y in zip(predictions, y_test)]) / len(predictions) * 100
print(f'Prediction accuracy {acc}')
```
